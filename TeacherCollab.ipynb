# Project Overview
1. Project aims to explore the types of challenges that teachers face in their collaborative planning sessions.
2. Data includes 12 responses from a single school regarding Levine's 5 Challenges of a PLC.
  a. Each question was on a 7-point scale, 1 being "highly disagree" and 7 being "highly agree"

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
```

```python
# Brining in dataset
data = pd.read_csv('/content/TLT Collaboration at PCA (Responses) - Form Responses 1.csv')
df = pd.DataFrame(data)
print(df)
```

```python
# Getting info about dataset for cleaning
df.info()
```

# Questions to explore
1. What does staff think about how PLC sessions influence staff and stakeholders?
2. How does staff perceive themselves as participants in the PLC process?
3. How does staff perceive PLC sessions as impacting their practice?

## *What does staff think about how PLC sessions influence staff and stakeholders?*

```python
# Getting the actual column names from the DataFrame
print(df.columns)

df_staff = df[['TLT sessions at PCA provide opportunities to explore diverse topics deeply and with purpose. ', 'TLT sessions at PCA provide opportunities for staff to reflect on their practice with other staff.', 'TLT sessions at PCA provide opportunities to engage and collaborate with others outside your content area. ', 'TLT sessions at PCA provide opportunities to engage and collaborate with stakeholders outside of PCA. ']]

# Sorting into the columns needed to investigate the first question
```

```python
df_staff
```

```python
df_staff.describe()
```

```python
# Visualizing the mean scores
df_staff.mean().plot.barh()
```

# Initial analysis
* When asked regarding agreement if **"TLT sessions at PCA provide opportunities to engage and collaborate with others outside your content area"**, the mean rating was 6.25, the highest mean. This suggests that most teachers surveyed agree that PLC sessions allow for collaboration outside content areas.

* When asked whether **"TLT sessions at PCA provide opportunities to engage and collaborate with stakeholders outside of PCA"** the mean score was 4.17, the lowest mean. This suggests that most teachers surveyed are neutral regarding whether PLC allow for collaboration with stakeholders outside their school.

## *2. How does staff perceive themselves as participants in the PLC process?*

```python
df_participants = df[['TLT sessions at PCA provide opportunities for all staff to participate in TLT, from leading sessions or discussions to participating in active learning. ',
       'TLT sessions at PCA maximize time and resources to support optimal learning of staff. ',
       'TLT sessions at PCA provide opportunities to have difficult, but necessary conversations.',
       'TLT sessions at PCA provide a sense of trust to me as a staff member. ']]
df_participants
```

```python
# Focusing on mean ratings
df_participants.mean()
```

# Initial analysis
* When asked whether: "TLT sessions at PCA provide opportunities for all staff to participate in TLT, from leading sessions or discussions to participating in active learning", the mean agreement score was 6.4. This suggests teachers feel very active in the PLC process.
* When asked whether: "TLT sessions at PCA provide opportunities to have difficult, but necessary conversations", the mean agreement score was 4.9, suggesting that teachers are neutral regarding whether PLC allow for staff to have necessary conversations to advance.  

## *3. How does staff perceive PLC sessions as impacting their practice?*

```python
df_impact = df[['TLT sessions at PCA have protocols for grade-level teams and other staff to collaborate on practice and share what works best for learners. ',
       'TLT sessions at PCA help me as an educator/professional serve the needs of learners.']]

df_impact
```

```python
# Focusing on mean ratings
df_impact.mean()
```

# Initial analysis
* Given that teachers surveyed gave mean scores higher than 6 for both measures, this suggests that PLC members feel that these sessions help inform their practice.

## Discussion
Based on the survey results, it seems that the following challenges to PLC collaboration may be present with this sample:
* **This PLC may not have opportunities for collaboration with stakeholders outside their school**. This challenge may present has frustruation with not having outside feedback, being unsure if teaching and learning is preparing students for college/career, or overall boredom.
* **This PLC may not have the opportunities for staff to have necessary conversations to advance.** This challenge may present itself as stagnation in discussions leading to stalled designs, for conflict to 'fester' and impact teaching and learning, or other blocked reasons for collaboration.
